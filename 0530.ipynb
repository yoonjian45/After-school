## 데이터 전처리 -> 데이터 세트 분리 -> 모델 학습 및 점증 평가 -> 예측 수행 -> 예측 평가

#### 원-핫 인코딩은 단어 집합의 크기만큼 벡터 차원을 가지고, 해당되는 단어가 있으면 1, 그렇지 않으면 0으로 표현하는 방식입니다

```python
import numpy as np

array = np.array([1,2,2,3,3,4,4,4,4])

unique_values, counts = np.unique(array,return_counts = True)

for values, counts in zip(unique_values,counts):
    print(f'value {values} appears {counts} times')
```
